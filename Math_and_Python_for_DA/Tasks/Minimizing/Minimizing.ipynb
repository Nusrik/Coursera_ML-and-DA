```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import optimize
import math
```

```python
def f(x):
    return math.sin(x / 5.) * math.exp(x / 10.) + 5 * math.exp(-x / 2.)
```

```python
x = np.arange(1, 31, 0.01)
```

```python
y = list(map(f, x))
```

```python
plt.plot(x, y)
plt.show()
```

```python
first = optimize.minimize(f, 2, method="BFGS")
```

```python
second = optimize.minimize(f, 30, method="BFGS")
```

```python
with open("ans-1.txt","w") as file_obj:
    ans = str("%.2f"%first.fun) + " " + str("%.2f"%second.fun)
    file_obj.write(ans)
```
